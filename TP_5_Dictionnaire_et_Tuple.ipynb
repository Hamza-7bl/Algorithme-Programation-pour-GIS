Exercice 3 : Distance simplifiée 

```python

def distance(v1, v2): 
    return abs(v1[0] - v2[0]) + abs(v1[1] - v2[1])

```

```python
#Appelez cette fonction pour afficher la distance entre Rabat et Casablanca.
d_R_C=distance(villes["Rabat"], villes["Casablanca"])
print("la distance entre Rabat et Casablanca est :",d_R_C)
```

```python
#Affichez la ville la plus éloignée de Rabat (en utilisant une boucle).
d=0
for i in villes.keys():
    if distance(villes["Rabat"], villes[i])>d:
        d = distance(villes["Rabat"], villes[i])
        key = i
print(f"la ville la plus eloigner de rabat est {key} de distance {d}")
```

Exercice 6 : Extraction des villes par pays 

```python
villes_info = { 
"Rabat": {"latitude": 34.020882, "longitude": -6.841650, "pays": "Maroc"}, 
"Casablanca": {"latitude": 33.573110, "longitude": -7.589843, "pays": "Maroc"}, 
"Tanger": {"latitude": 35.759465, "longitude": -5.833954, "pays": "Maroc"}, 
"Paris": {"latitude": 48.8566, "longitude": 2.3522, "pays": "France"} 
}
```

```python
#Créez une fonction villes_par_pays(dictionnaire, nom_pays) qui retourne une liste des noms de villes situées dans ce pays. 
def villes_par_pays(dictionnaire, nom_pays):
    l=[]
    for i in dictionnaire:
        if nom_pays in dictionnaire[i].values():
            l.append(i)
    print(f"les villes de {nom_pays} : {l}")
```

```python
#Testez-la pour le pays "Maroc".
villes_par_pays(villes_info, "Maroc")
```

```python
#Créez une fonction villes_par_pays(dictionnaire, nom_pays) qui retourne une liste des noms de villes situées dans ce pays. 
def villes_par_pays2(dictionnaire, nom_pays):
    l=[]
    for i in dictionnaire:
        if nom_pays == dictionnaire[i]["pays"]:
            l.append(i)
    print(f"les villes de {nom_pays} : {l}")

villes_par_pays2(villes_info, "France")
```
