Exercice 1 : Représentation et affichage des données spatiales

```python
import math
#Affichage des points
def afficher_points(points):
    for i in range(len(points)) :
        print(f"{points[i][0]} | lattiude : {points[i][1]} | longitude : {points[i][2]} ")


#Calcul de distance entre deux points
def distancee():
    lat1 = float(input("donner lat1 :"))
    lon1 = float(input("donner lon2 :"))
    lat2 = float(input("donner lat1 :"))
    lon2 = float(input("donner lon1 :"))
    return math.sqrt( (lat2-lat1)**2 + (lon2-lon1)**2 )


#Calcul de distance entre deux villes
def distance(p1,p2):
    return math.sqrt( (p2[1]-p1[1])**2 + (p2[2]-p1[2])**2 )


# Distance moyenne
def distance_moyen(points):
    list_d = []
    for i in range(1,len(points)):

        list_d.append(distance(points[0],points[i]))
    distance_moyen = sum(list_d)/(len(points)-1)
    print(list_d)
    print("distance_moyen est : ",distance_moyen)



#Recherche du client le plus proche du dépôt
def distance_plus_proche(points):
    list_d = []
    for i in range(1,len(points)):
        list_d.append(distance(points[0],points[i]))
    distance_pproche = min(list_d)
    p = list_d.index(distance_pproche)
    print(f"le client le plus proche du depot est {points[p+1][0]} de distanca {distance_pproche} degree")



#Points dans une zone
def points_dans_zone(points, lat_min, lat_max, lon_min, lon_max):
    points_zone = []
    p_zone = []
    for i in points :
        if (i[1]<= lat_max and i[1] >= lat_min) and (i[2]<= lon_max and i[2] >= lon_min):
            points_zone.append(i)
    if points_zone != [] :
        for i in range(len(points_zone)) :
            p_zone.append(points_zone[i][0])
        print("les point dans la zone est ",p_zone)
    else :
        print("il n'ya pas de points dans la zone demander ")

```

```python
points = [["Depot", 33.5731, -7.5898], ["Client A", 33.5899, -7.6039],["Client B", 33.5735, -7.6467], ["Client C", 33.6100, -7.5300] ]
```

1. Affiche tous les points.

```python
print(afficher_points(points))
```

2. Calcule et affiche la distance entre le dépôt et chaque point.

```python
d = []
for i in range(1,len(points)) :
    d.append(distance(points[0],points[i]))
    print(f"la distance entre depot et {points[i][0]} est : {d[i-1]} degree")
```

3. Indique le client le plus proche du dépôt.

```python
print(distance_plus_proche(points))
```

4. Affiche les clients situés dans la zone géographique donnée.

```python
print(points_dans_zone(points,  33.57, 33.60, -7.65, -7.58))
```
