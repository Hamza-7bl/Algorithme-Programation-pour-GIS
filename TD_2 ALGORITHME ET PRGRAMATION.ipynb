```python
# Exercice 7 : Moyenne et filtrage
# Soit la liste des notes suivantes :
# notes = [12, 8, 17, 5, 14, 9, 18]
# 1. Calculer et afficher la moyenne des notes.
# 2. Créer une nouvelle liste contenant uniquement les notes supérieures ou égales à 10.
# 3. Afficher cette nouvelle liste.

# 1. Calculer et afficher la moyenne des notes.
notes = [12, 8, 17, 5, 14, 9, 18]
somme = 0
moyenne = 0
for i in notes:
  somme +=i
  moyenne = somme/len(notes)

print(round(moyenne,2))

# 2. Créer une nouvelle liste contenant uniquement les notes supérieures ou égales à 10.
list_sup = []
for i in notes:
  if i >= 10:
    list_sup.append(i)

# 3. Afficher cette nouvelle liste.
print(list_sup)
```
