```python
#Exercice 2 - Convertisseur de devises
while True:
    # Saisie du montant en dirhams
    montant_dh = float(input("Entrez le montant en dirhams : "))
    
    # Taux de change
    taux_euro = 11
    taux_dollar = 10
    
    # Conversion
    montant_euro = montant_dh / taux_euro
    montant_dollar = montant_dh / taux_dollar
    
    # Affichage des résultats
    print(f"{montant_dh} DH = {montant_euro:.2f} €")
    print(f"{montant_dh} DH = {montant_dollar:.2f} $")
    
    # Demander si l'utilisateur veut recommencer
    continuer = input("Voulez-vous faire une autre conversion ? (oui/non) : ")
    if continuer.lower() != "oui":
        print("Au revoir !")
        break
```

```python
#Exercice 3 - Menu interactif
def calculatrice():
    print("\n--- Calculatrice ---")
    a = float(input("Entrez le premier nombre : "))
    b = float(input("Entrez le deuxième nombre : "))
    print(f"{a} + {b} = {a + b}")
    print(f"{a} - {b} = {a - b}")
    print(f"{a} × {b} = {a * b}")
    if b != 0:
        print(f"{a} ÷ {b} = {a / b}")

def conversion_temps():
    print("\n--- Conversion temps ---")
    secondes = int(input("Entrez le nombre de secondes : "))
    heures = secondes // 3600
    minutes = (secondes % 3600) // 60
    sec_restantes = secondes % 60
    print(f"{secondes} secondes = {heures}h {minutes}min {sec_restantes}s")

def afficher_suite():
    print("\n--- Suite numérique ---")
    n = int(input("Combien de termes afficher ? "))
    prmier_terme =1
    for i in range(n):
        print(prmier_terme)
        prmier_terme *=2
    print()

# Programme principal
while True :
    print("\n=== MENU PRINCIPAL ===")
    print("1. Calculatrice")
    print("2. Conversion temps")
    print("3. Afficher une suite")
    print("4. Quitter")
    
    choix = input("Choisissez une option (1-4) : ")
    
    if choix == "1":
        calculatrice()
    elif choix == "2":
        conversion_temps()
    elif choix == "3":
        afficher_suite()
    elif choix == "4":
        print("Au revoir !")
        break
    else:
        print("Option invalide !")
```
